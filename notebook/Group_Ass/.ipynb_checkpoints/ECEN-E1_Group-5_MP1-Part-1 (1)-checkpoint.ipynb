# To start....

First, a list of numeric and Non-Numeric Features was created and then the numeric features was plotted using regression plot.

<h2>Creating List of Numeric and Non-Numeric Features.</h2>

The goal of this block was to make a separate list of numeric and non numeric features. A for loop was used to scan all the columns present in the dataset. The column value was checked if it contains numeric or non numeric value. The output of the two was appended to a separate list.

```python
target = 'Weight' # Output Column to be predicted

#Create List of Features
numeric_features = [] # List of the Numeric Features
non_numeric_features = [] # List of the Non Numeric Features

# Scan all of the columns of the dataset
for col in dataset.columns:
    
    if col != target: # Check if the column scanned is the Target Output. Disregard the column if it is the Target Column

        # Check if the column is numeric feature
        if pd.api.types.is_numeric_dtype(dataset[col]):
           numeric_features.append(col) # Append numeric to numeric feature list
        
        else: # A non numeric feature
            non_numeric_features.append(col)
            continue

#Print the results
print(f"Numeric Features of Dataset: ",numeric_features) 
print(f"Non Numeric Features of Dataset: ",non_numeric_features)
```

<h2>Visualization of Linearity of the Numeric Features to the Target Output</h2>

The goal of this section is to view the linearity of the selected feature in respect to its target output using regression plotting

```python
# Get number of need subplots
n_features = len(numeric_features)

# Set the number of rows and columns subplot grid
n_cols = 2  # 2_Column Subplot
n_rows = (n_features + n_cols - 1) // n_cols  # Compute number of rows to fit all the features

# Create the figure with adjusted height
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 4), constrained_layout=False)

# Flatten the graph
axes = axes.flatten()

# Loop through features and plot
for idx, feature in enumerate(numeric_features):
    sns.regplot(x=feature, y=target, data=dataset, ax=axes[idx],
                color='lightsalmon',
                line_kws={"color": "green", "linewidth": 1.5,},
                marker = "+")
    axes[idx].set_title(f'{feature} vs {target}'.upper(), fontsize=12, fontweight='bold')

# Hide any empty plots from extra subplot grid
for idx in range(n_features, len(axes)):
    axes[idx].axis('off')  
    
# Adjust spacing between subplots
plt.subplots_adjust(hspace=0.5, wspace=0.3)  # Increase vertical and horizontal space

# Show the plots
plt.show()
```

A visual representation of correlation value was not enough to prove if the feature really affects the target output value. 
To further strenghten the conclusion, a Statistical Analysis was used to analyze the feature's p_value to the target value of the numerical features.

<h3>Verifying The Correlation Between Numeric Features and The Target Output using Statistical Analysis with Pearsonr</h3>

```python
from scipy.stats import pearsonr


correlation_table = {} 

# Get the Correalation Coefficient and the P-Value of all Numeric Features
for feature in numeric_features:
    corr, p_value = pearsonr(dataset[feature], dataset[target]) # Set value corr and p_value for correlation value and p_value
    correlation_table[feature] = {'Correlation Coeff': corr, 'P-value': p_value} # Summarize the corr value and p_value to a list

#Print the Correlation Coefficient and the P-Value
result = pd.DataFrame(correlation_table).T # Get the transpose of the list from correlation_table
print(result)
```

<h4>Conclusion:</h4>

All features were observed to be significant in terms of the P-value. However, upon observing the Perason's Correlation Coefficient of each features, the numerical features that can only be considered as a prospective factor in preticting the output variable (weight) are the **Age, Height,** and **CH2O (Amount of Daily Consumption of Water)**.

to the target output value

<h2>Visualization of Correlation of the Non-Numeric Features to the Target Output Using Boxplot</h2>

```python
# Get number of need subplots
n_features = len(non_numeric_features)

# Set the number of rows and columns of subplot grid
n_cols = 2  # 2_Column Subplot
n_rows = (n_features + n_cols - 1) // n_cols  # Compute number of rows to fit all the features

# Create the figure with adjusted height
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 6, n_rows * 5), constrained_layout=False)

# Flatten axes for easier looping
axes = axes.flatten()

# Loop through features and plot
for idx, feature in enumerate(non_numeric_features):
    sns.boxplot(x=feature, y=target, data=dataset, ax=axes[idx], palette='viridis')
    
   
    axes[idx].set_title(f'{feature} vs {target}'.upper(), fontsize=12, fontweight='bold')
    xticks = axes[idx].get_xticklabels()
    
    if any(len(label.get_text()) > 9 for label in xticks): # Check if the labels are too long
        axes[idx].tick_params(axis='x', rotation=45)  # Rotate labels if they are long
    
    axes[idx].get_legend().remove() if axes[idx].get_legend() else None # Remove the Legends of the subplots
    
# Hide any empty plots from extra subplot grid
for idx in range(n_features, len(axes)):
    axes[idx].axis('off')  
    
# Adjust spacing between subplots
plt.subplots_adjust(hspace=0.8, wspace=0.3) # Increase vertical and horizontal space

# Show the plots
plt.show()
```

<h2>Visualizing the Distribution of Unique Values in Non-numeric Features to Assess their Balance.</h2>

```python
# Set up the grid
n_features = len(non_numeric_features)

# Set the number of rows and columns
n_cols = 2  # 2_Column Subplot
n_rows = (n_features + n_cols - 1) // n_cols  # Compute number of rows to fit all the features

# Create the figure with adjusted height
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 7, n_rows * 7), constrained_layout=False)

# Flatten axes for easier looping
axes = axes.flatten()

# Loop through features and plot
for idx, feature in enumerate(non_numeric_features):
    ax = sns.histplot(x=feature, data=dataset, ax=axes[idx], legend=False, hue=feature)

    for p in ax.patches:
        count = int(p.get_height())

        if count > 0:  
            ax.annotate(f'{count}', 
                        (p.get_x() + p.get_width() / 2., p.get_height()), 
                        xytext=(0, 1), 
                        textcoords='offset points',
                        ha='center', va='bottom',
                        fontsize=9,)
            
    axes[idx].set_title(f'{feature} vs {target}'.upper(), fontsize=12, fontweight='bold')

# Adjust spacing between subplots
plt.subplots_adjust(hspace=0.3, wspace=0.3)  # Increase vertical and horizontal space
plt.xticks(rotation=45)

# Show the plots
plt.show()
```

All non-numerical features were re-evaluated using Analysis of Variance (ANOVA) to confirm whether these non-numerical features are truly significant in predicting the output.

## Calculating the Analysis of Variance (ANOVA) of the Non-Numerical Features in Relation to the Target Values. 

<h4>Importing necessary Libraries for ANOVA calculation</h4>

```python
from scipy import stats
```

### Performing ANOVA for each feature in relation to the Target.

```python
Anova_table = {}

for feature in non_numeric_features:
    # Group data by 'Frequency'
    grouped = dataset.groupby(feature)

    # Perform ANOVA
    f_stat, p_value = stats.f_oneway(*(grouped.get_group(group)[target] for group in grouped.groups))
    Anova_table[feature] = {'F-Statistics': f_stat, 'P-value': p_value}


result = pd.DataFrame(Anova_table).T
print(result)
```

To determine the significance of a feature as a predicting factor of the target, the F-statistics should be at a large positive value, while its P-value should be at a value closest to zero (0) as possible.

Based on the value of the F-Statistics, the features that would be a possible predicting factor are as follows:

1. Gender
2. family_history_with_overweight
3. FAVC
4. CAEC
5. SCC
6. CALC
7. NObeyesdad


### Calculating the ANOVA between the Unique Values of the Non-Numerical Features

```python
import itertools

sel_Anova_feature = {'Gender','family_history_with_overweight', 'FAVC', 'CAEC',
                     'SCC', 'CALC', 'NObeyesdad'}

sel_Anova_table = {}

for feature in sel_Anova_feature:
    # Get unique categories from the categorical feature
    unique_categories = dataset[feature].unique()



    # Loop through each pair of categories and perform ANOVA
    for cat1, cat2 in itertools.combinations(unique_categories, 2):
        # Extract the 'Value' for each category
        group1 = dataset[dataset[feature] == cat1][target]
        group2 = dataset[dataset[feature] == cat2][target]
        
        # Perform ANOVA
        f_stat, p_value = stats.f_oneway(group1, group2)

        # Store the p-value for this pair of categories
        sel_Anova_table[f"<{feature}> {cat1} vs {cat2} "] = {'F-Statistics': f_stat, 'P-value': p_value}
    

# Display p-values and F-statistics between each pair of categories
result = pd.DataFrame(sel_Anova_table).T
print(result)
```

Upon observation of the selected potential features, non-numerical features **CAEC** and **CALC** produced a value closer to zero (0), which indicated that it is not a predicting factor of the output. These features were then removed.
