### Filling the Missing Data for Categorical Valued Columns

To fill the missing data for Categorical Valued columns, the most frequent value within the column will be inserted.

#### List all the Categorical Valued Columns

```python
categorical_columns = dataset_new[["Gender", "family_history_with_overweight", "FAVC", 
                                   "CAEC", "SMOKE", "SCC", "CALC", "MTRANS"]]
```

#### Replace "NaN" using Simple Imputer

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')

dataset_new[["Gender", "family_history_with_overweight", "FAVC", "CAEC", "SMOKE", "SCC", "CALC", "MTRANS"]] = imputer.fit_transform(categorical_columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
filename = 'ObesityDataset_Cleaned.csv'
dataset = pd.read_csv(filename)
```

```python
dataset.head(5)
```

```python
print(dataset.dtypes)
```

```python
corr_matrix = dataset.corr(numeric_only=True)

# Set the figure size and improve appearance
plt.figure(figsize=(15, 15))

# Create a heatmap with annotations
sns.heatmap(corr_matrix, annot=True, cmap='hsv', center=0, fmt='.2f',
            annot_kws={'size': 12}, linewidths=0.5, cbar_kws={'shrink': 0.8})

# Set the title
plt.title('Correlation Matrix', fontsize=16)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
target = 'NObeyesdad'
numeric_features = []
non_numeric_features = []

for col in dataset.columns:
    if col != target:
        if pd.api.types.is_numeric_dtype(dataset[col]):
           numeric_features.append(col)
        else:
            non_numeric_features.append(col)
            print(f"Skipping non-numeric column: {col}")
            continue

print(numeric_features)
```

```python

# Set up the grid
n_features = len(numeric_features)

# Set the number of rows and columns
n_cols = 2  # Fixed to 2 columns
n_rows = n_features // n_cols  # Compute rows to fit all features

# Create the figure with adjusted height
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 4), constrained_layout=False)

# Flatten axes for easier looping
axes = axes.flatten()

# Loop through features and plot
for idx, feature in enumerate(numeric_features):
    sns.regplot(x=feature, y=target, data=dataset, ax=axes[idx],
                scatter_kws={'s': 50, 'alpha': 0.7},  # transparency for scatter points
                line_kws={'color': 'green', 'linewidth': 2},
                fit_reg=True,  # Fit a regression line
                scatter=True)
    axes[idx].set_title(f'{feature} vs {target}'.upper(), fontsize=12, fontweight='bold')

# Adjust spacing between subplots
plt.subplots_adjust(hspace=0.5, wspace=0.3)  # Increase vertical and horizontal space

# Show the plots
plt.show()
```

```python
from scipy.stats import pearsonr


correlation_table = {}

for feature in numeric_features:
    corr, p_value = pearsonr(dataset[feature], dataset[target])
    correlation_table[feature] = {'Correlation': corr, 'P-value': p_value}

result = pd.DataFrame(correlation_table).T
print(result)
```

```python
# Set up the grid
n_features = len(non_numeric_features)

# Set the number of rows and columns
n_cols = 2  # Fixed to 2 columns
n_rows = (n_features // n_cols)   # Compute rows to fit all features

# Create the figure with adjusted height
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 6, n_rows * 5), constrained_layout=False)

# Flatten axes for easier looping
axes = axes.flatten()

# Loop through features and plot
for idx, feature in enumerate(non_numeric_features):
    sns.boxplot(x=feature, y=target, data=dataset, ax=axes[idx], hue=feature, palette='viridis')
    axes[idx].set_title(f'{feature} vs {target}'.upper(), fontsize=12, fontweight='bold')

# Adjust spacing between subplots
plt.subplots_adjust(hspace=0.4, wspace=0.3)  # Increase vertical and horizontal space
plt.xticks(rotation=45)

# Show the plots
plt.show()
```
