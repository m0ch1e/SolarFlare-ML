<h2>Verifying The Correlation Between Numeric Features and The Target Output using Statistical Analysis with Pearsonr</h2>

<h3>Creating List of Numeric and Non-Numeric Features.</h3>

<p>The goal of this block was to make a separate list of numeric and non numeric features. A for loop was used to scan all the columns present in the dataset. The column value was checked if it contains numeric or non numeric value. The output of the two was appended to a separate list.</p>

```python
#Create List of Features
numeric_features = [] # List of the Numeric Features
non_numeric_features = [] # List of the Non Numeric Features

# Scan all of the columns of the dataset
for col in dataset.columns:
    
    if col != target: # Check if the column scanned is the Target Output. Disregard the column if it is the Target Column

        # Check if the column is numeric feature
        if pd.api.types.is_numeric_dtype(dataset[col]):
           numeric_features.append(col) # Append numeric to numeric feature list
        
        else: # A non numeric feature
            non_numeric_features.append(col)
            continue

#Print the results
print(f"Numeric Features of Dataset: ",numeric_features) 
print(f"Non Numeric Features of Dataset: ",non_numeric_features)
```

```python
from scipy.stats import pearsonr


correlation_table = {} 

# Get the Correalation Coefficient and the P-Value of all Numeric Features
for feature in numeric_features:
    corr, p_value = pearsonr(dataset[feature], dataset[target]) # Set value corr and p_value for correlation value and p_value
    correlation_table[feature] = {'Correlation Coeff': corr, 'P-value': p_value} # Summarize the corr value and p_value to a list

#Print the Correlation Coefficient and the P-Value
result = pd.DataFrame(correlation_table).T # Get the transpose of the list from correlation_table
print(result)
```

<p><strong>Conlcusion:</strong> 
    <p>[1] Based on the result of the pearson correlation, the features does not exhibit dependence to the target, thus, the approach to classify the target is to use classification method.</p>
    <p>[2] MEANSHR, TOTFX, Mhis1d, Mdec, Edec, NOAA, HARP shows low correlation to the target </p>

```python
from scipy.stats import kendalltau

correlation_table = {} 

# Get the Correalation Coefficient and the P-Value of all Numeric Features
for feature in numeric_features:
    tau, p_value = kendalltau(dataset[feature], dataset[target]) # Set value corr and p_value for correlation value and p_value
    correlation_table[feature] = {'Tau': tau, 'P-value': p_value} # Summarize the corr value and p_value to a list

#Print the Correlation Coefficient and the P-Value
result = pd.DataFrame(correlation_table).T # Get the transpose of the list from correlation_table
print(result)
```

```python
from sklearn.inspection import plot_partial_dependence
from sklearn.ensemble import RandomForestRegressor

# Fit a random forest model
rf = RandomForestRegressor()
rf.fit(dataset[['feature']], dataset['target'])

# Generate partial dependence plot
plot_partial_dependence(rf, dataset[['feature']], ['feature'])
plt.show()

# Get the Correalation Coefficient and the P-Value of all Numeric Features
for feature in numeric_features:
    tau, p_value = kendalltau(dataset[feature], dataset[target]) # Set value corr and p_value for correlation value and p_value
    correlation_table[feature] = {'Tau': tau, 'P-value': p_value} # Summarize the corr value and p_value to a list

#Print the Correlation Coefficient and the P-Value
result = pd.DataFrame(correlation_table).T # Get the transpose of the list from correlation_table
print(result)
```

```python
from sklearn.inspection import PartialDependenceDisplay
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt

# Fit a random forest model
rf = RandomForestRegressor()
rf.fit(dataset[numeric_features], dataset[target])

# Generate partial dependence plots for all features
PartialDependenceDisplay.from_estimator(
    rf, dataset[numeric_features], numeric_features, 
    grid_resolution=50
)

# Adjust the figure size to prevent overlap
plt.figure(figsize=(20, 15))  # Increase the figure size

# Rotate x-labels and/or y-labels for better visibility
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.yticks(rotation=45)  # Optionally rotate y-axis labels if needed

# Adjust layout with more spacing between subplots
plt.subplots_adjust(hspace=0.5, wspace=3)  # Increase hspace and wspace to prevent overlap
# Show the plot
plt.tight_layout()  # Adjust layout to avoid overlapping labels
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor



# Fit the Decision Tree model
tree = DecisionTreeRegressor(max_depth=3)
tree.fit(dataset[numeric_features], dataset[target])

# Display feature importance for each feature
feature_importances = tree.feature_importances_
for feature, importance in zip(numeric_features, feature_importances):
    print(f"Feature: {feature}, Importance: {importance:.4f}")
```

```python
# Get number of need subplots
n_features = len(numeric_features)

# Set the number of rows and columns subplot grid
n_cols = 2  # 2_Column Subplot
n_rows = (n_features + n_cols - 1) // n_cols  # Compute number of rows to fit all the features

# Create the figure with adjusted height
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 4), constrained_layout=False)

# Flatten the graph
axes = axes.flatten()

# Loop through features and plot
for idx, feature in enumerate(numeric_features):
    sns.regplot(x=feature, y=target, data=dataset, ax=axes[idx],
                color='lightsalmon',
                line_kws={"color": "green", "linewidth": 1.5,},
                marker = "+")
    axes[idx].set_title(f'{feature} vs {target}'.upper(), fontsize=12, fontweight='bold')

# Hide any empty plots from extra subplot grid
for idx in range(n_features, len(axes)):
    axes[idx].axis('off')  
    
# Adjust spacing between subplots
plt.subplots_adjust(hspace=0.5, wspace=0.3)  # Increase vertical and horizontal space

# Show the plots
plt.show()
```

```python
import pandas as pd


df = pd.DataFrame(dataset["flare"])

# Specify the column(s) to analyze
columns_to_analyze = df.columns  # Replace with specific columns, e.g., ['col1', 'col2']

# Prepare a list to store unique counts
unique_counts_list = []

# Count unique values for each column
for col in columns_to_analyze:
    unique_count = df[col].nunique(dropna=False)  # Count unique values, including NaN
    unique_counts_list.append({'Column': col, 'Unique_Count': unique_count})

# Convert the result to a DataFrame
unique_counts_df = pd.DataFrame(unique_counts_list)

# Export the unique counts to CSV
unique_counts_df.to_csv('unique_counts_output.csv', index=False)

print("Unique value counts have been successfully exported to 'unique_counts_output.csv'.")
```

```python
import pandas as pd

df = pd.DataFrame(dataset["flare"])

# Prepare lists for results
output_list = []  # To store value counts
unique_counts_list = []  # To store unique counts

# Loop through all columns
for col in df.columns:
    # Get value counts for the column
    value_counts = df[col].value_counts(dropna=False)
    
    # Add value counts to the output list
    temp_df = value_counts.reset_index()
    temp_df.columns = ['Value', 'Count']  # Rename columns for clarity
    temp_df['Column'] = col  # Add column name
    output_list.append(temp_df)
    
    # Add unique count for the column
    unique_counts_list.append({'Column': col, 'Unique_Count': df[col].nunique(dropna=False)})

# Combine value counts into a single DataFrame
value_counts_df = pd.concat(output_list, ignore_index=True)

# Reorder columns for clarity
value_counts_df = value_counts_df[['Column', 'Value', 'Count']]

# Convert unique counts to a DataFrame
unique_counts_df = pd.DataFrame(unique_counts_list)

# Merge both DataFrames (if needed) or export separately
result_df = pd.merge(value_counts_df, unique_counts_df, on='Column', how='left')

# Export to CSV
result_df.to_csv('value_counts_and_unique_output.csv', index=False)

print("Value counts and unique counts have been exported to 'value_counts_and_unique_output.csv'.")
```
